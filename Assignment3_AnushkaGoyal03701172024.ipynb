```python
def printNum(n):
  for i in range(1,n+1):
    print(i)

print(printNum(10))
```

```python
def calcSum(n):
  sum = 0;
  for i in range(1,n+1):
    sum += i;
  print(sum)
print(calcSum(10))
```

```python
def reverseNum(num):
  reversed_num = 0
  temp = abs(num)
  while temp > 0:
    digit = temp % 10
    reversed_num = (reversed_num*10)+digit
    temp = temp // 10

  if num < 0:
    reversed_num = -reversed_num
  print(reversed_num)
print(reverseNum(12345))
```

```python
def checkPalindrome(num):
  temp = num
  rev_num = 0
  while temp > 0 :
    digit = temp % 10
    rev_num = (rev_num)*10 + digit
    temp = temp // 10

  if(rev_num == num):
    print("They are palindrome !!")
  else:
    print("They are not palindrome")
checkPalindrome(12321)
```

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

terms = 10
# Loop is required to print the entire sequence up to n
for i in range(terms):
    print(fibonacci_recursive(i), end=" ")
```
