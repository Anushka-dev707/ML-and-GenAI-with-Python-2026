QUESTION 2 - TO PRINT FACTORIAL OF THE NUMBER


```python
num = int(input("Enter the number : "))
factorial = 1;
for i in range (1,num+1):
  factorial *= i;
print(factorial)
```

#QUESTION 3 - PRINT FIBONACCI SERIES


```python
def print_fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        print(a, end=" ")
        a, b = b, a + b

# Print the first 10 terms
print_fibonacci(10)
```
